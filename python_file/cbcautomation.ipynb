```python
# CBC GRADING 
def cbc_report_generator(df): 
          
    # --- Automates CBC grading, comments and general remarks for a student dataframe of ---; 
    
# 1.CLEANING 
 # fill missing scores with 0 
 df = df.fillna(0)  

 #2.CONVERTING FLOATS TO INT 
 subjectcolumns = [col for col in df.columns if col not in ['Position','NAME','TOTAL']]
 df[subjectcolumns] = df[subjectcolumns].astype(int)  
    
 #3. GRADING     
 def get_grade(score):
    if score >= 89:    return 'EE1'
    elif score >= 74:  return 'EE2' 
    elif score >= 57:  return 'ME1' 
    elif score >= 40:  return 'ME2'
    elif score >= 30:  return 'AE1' 
    elif score >= 20:  return 'AE2'
    elif score >= 10:  return 'BE1' 
    elif score >= 1:   return 'BE2'
    else:              return 'ABSENT' 
 # specifying columns for grading 
 for col in subjectcolumns: 
    df[col] = df[col].apply(get_grade) 

#4. COMMENTS SECTION 
# generating comments
 def get_comment(grade):
    if   'EE' in grade:  return "Exceeding Expectations"
    elif 'ME' in grade:  return "Meeting Expectations"
    elif 'AE' in grade:  return "Approcahing Expectations"
    elif 'BE' in grade:  return 'Below Expectations'
    else :               return "ABSENT" 

# adding a comment column based on learning area 
 for col in subjectcolumns:
    comment_col = col + "_COMMENT"
    df[comment_col] = df[col].apply(get_comment) 
   
#5.  ADDING A GENERAL REMARKS COLUMN 
 def generalcomment(position):    
    if  1 <= position <= 10:                         #SAME AS; If position >= 1 and position <= 10 
        return 'Good work, Keep working hard.'
    elif  11 <= position <= 25: 
        return 'You did well, put more effort in your weak learning areas.'
    elif  26 <= position <= 36:
        return 'You can do better, work harder.'
    elif   37 <= position <= 45: 
        return " There's room for improvement, put more effort."
    else:
        return 'Pull up your socks! You can do better.' 
             
# adding the general remarks column 
 df['GENERAL_COMMENT'] = df['Position'].apply(generalcomment)  

#6. ADD DATES DEPENDING ON TERM DATES
# df['CLOSING_DATE'] = 'YYYY-MM-DD'
# df['OPENING_DATE'] = 'YYYY-MM-DD' 

 return df 

# --- HOW TO USE IT ---; 
# processed_report = cbc_report_generator(df)
```
